```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 18))
```

```python
# open image as numpy
img = cv2.imread('my_favor.jpg')
```

```python
# image attribute
print (f'img type : {type(img)}')
print (f'img shape : {img.shape}')
print (f'img values : {np.unique(img)}')
print (f'img element dtype : {img.dtype}')
```

```python
plt.imshow(img)
```

```python
# basic operations

# color transformation
# bgr to rgb
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# rgb = np.flip(img, axis=-1)

# bgr to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow(gray, cmap='gray')

# binarizions
ret, gray = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
print (np.unique(gray))
plt.imshow(gray / 255., cmap='gray')
```

```python
# resize
# code: bicubic, linear, nearest, ...
img = cv2.resize(img, dsize=(40, 40), interpolation=cv2.INTER_AREA)
plt.imshow(img)

img = cv2.resize(img, dsize=None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)
plt.imshow(img)
```
