```python
# Re-define the function to calculate momentum for both players and to apply it to each point in the match
def calculate_p1_momentum_per_point(df):
    # Score Factor: using games and sets won
    df['Score Factor'] = 3 * (df['p1_sets'] - df['p2_sets']) + df['p1_games'] - df['p2_games']

    # Service Factor: using aces and double faults
    df['Service Factor'] = (df['p1_ace'] - df['p1_double_fault']) - (df['p2_ace'] - df['p2_double_fault'])

    # Break Factor: using break points won and missed
    # We add a small number to the denominator to prevent division by zero
    epsilon = 1e-6
    df['Break Factor'] = ((df['p1_break_pt_won'] - df['p1_break_pt_missed']) - 
                          (df['p2_break_pt_won'] - df['p2_break_pt_missed'])) / (df[['p1_break_pt', 'p2_break_pt']].sum(axis=1) + epsilon)

    # Rally Factor: using rally count and points won on long rallies
    # First, we determine the average rally count
    average_rally_count = df['rally_count'].mean()
    # Then, we calculate the ratio of points won on rallies longer than average for each player
    df['p1_long_rallies_won'] = df.apply(lambda row: 1 if (row['rally_count'] > average_rally_count and row['point_victor'] == 1) else 0, axis=1)
    df['p2_long_rallies_won'] = df.apply(lambda row: 1 if (row['rally_count'] > average_rally_count and row['point_victor'] == 2) else 0, axis=1)
    total_long_rallies = df['p1_long_rallies_won'].sum() + df['p2_long_rallies_won'].sum()
    df['Rally Factor'] = (df['p1_long_rallies_won'].sum() - df['p2_long_rallies_won'].sum()) / (total_long_rallies + epsilon)

    # Calculate total momentum for each point
    df['Momentum'] = df['Score Factor']*4 + df['Service Factor'] + df['Break Factor']*5 + df['Rally Factor']

    # Drop the intermediate columns used for the calculation
    df.drop(['Score Factor', 'Service Factor', 'Break Factor', 'p1_long_rallies_won', 'p2_long_rallies_won'], axis=1, inplace=True)

    return df

# Apply the function to calculate momentum per point
df_with_momentum = calculate_p1_momentum_per_point(df_1.copy())

# Fit the momentum to the match score trend
# For this, we could use a simple linear model but let's visualize the data first
import matplotlib.pyplot as plt

# Extract the cumulative score for each player
df_with_momentum['p1_cumulative_score'] = df_with_momentum["p1_points_won"]  # Multiply sets by 10 for visual distinction
df_with_momentum['p2_cumulative_score'] = df_with_momentum["p2_points_won"]
df_with_momentum['p1_cumulative_score_dis'] = df_with_momentum["p1_points_won"] - df_with_momentum["p2_points_won"]
# Plotting the momentum and the cumulative score over the points played
plt.figure(figsize=(14, 7))
plt.plot(df_with_momentum['point_no'], df_with_momentum['Momentum'], label='Momentum', color='blue', alpha=0.5)
#plt.plot(df_with_momentum['point_no'], df_with_momentum['p1_cumulative_score'], label='Player 1 Score', color='red', linestyle='--')
#plt.plot(df_with_momentum['point_no'], df_with_momentum['p2_cumulative_score'], label='Player 2 Score', color='green', linestyle='--')
plt.plot(df_with_momentum['point_no'], df_with_momentum['p1_cumulative_score_dis'], label='Points difference', color='brown', linestyle='-')
plt.xlabel('Point Number')
plt.ylabel('Momentum / Score')
plt.title('P1-Momentum and Score difference')
plt.legend()
plt.show()

# Output the dataframe with the new 'Momentum' column
df_with_momentum[['point_no', 'Momentum']].head()
```

```python
def calculate_p2_momentum_per_point(df):
    # Score Factor: using games and sets won
    df['Score Factor'] = df['p2_games'] * df['p2_sets'] - df['p1_games'] * df['p1_sets']

    # Service Factor: using aces and double faults
    df['Service Factor'] = (df['p2_ace'] - df['p2_double_fault']) - (df['p1_ace'] - df['p1_double_fault'])

    # Break Factor: using break points won and missed
    # We add a small number to the denominator to prevent division by zero
    epsilon = 1e-6
    df['Break Factor'] = ((df['p2_break_pt_won'] - df['p2_break_pt_missed']) - 
                          (df['p1_break_pt_won'] - df['p1_break_pt_missed'])) / (df[['p2_break_pt', 'p1_break_pt']].sum(axis=1) + epsilon)

    # Rally Factor: using rally count and points won on long rallies
    # First, we determine the average rally count
    average_rally_count = df['rally_count'].mean()
    # Then, we calculate the ratio of points won on rallies longer than average for each player
    df['p2_long_rallies_won'] = df.apply(lambda row: 1 if (row['rally_count'] > average_rally_count and row['point_victor'] == 2) else 0, axis=1)
    df['p1_long_rallies_won'] = df.apply(lambda row: 1 if (row['rally_count'] > average_rally_count and row['point_victor'] == 1) else 0, axis=1)
    total_long_rallies = df['p2_long_rallies_won'].sum() + df['p1_long_rallies_won'].sum()
    df['Rally Factor'] = (df['p2_long_rallies_won'].sum() - df['p1_long_rallies_won'].sum()) / (total_long_rallies + epsilon)

    # Calculate total momentum for each point
    df['Momentum'] = df['Score Factor']*4 + df['Service Factor'] + df['Break Factor']*5 + df['Rally Factor']

    # Drop the intermediate columns used for the calculation
    df.drop(['Score Factor', 'Service Factor', 'Break Factor', 'p2_long_rallies_won', 'p1_long_rallies_won'], axis=1, inplace=True)

    return df
```
